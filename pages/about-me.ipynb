# تعريف

> **" صدقني!  لن تعلق بمضيق، إذا قمت بمتابعة التحرك فقط "**

> *جوني أنكلش*



مرحباً, أختصاصي الأصلي هو هندسة الميكاترونيك , وهي هندسة حديثة (1970) تم إنشائها باليابان لأول مرة, ومنذ ذلك الحين كان مفهومها يتطور مع تقدم التقنيات ، ولذلك ظهرت لها عدة تعاريف.
ويمكن تعريفها على أنها التكامل الدقيق وليس مجرد الدمج بين الهندسات الالكترونية والميكانيكية ، ..

حصلت على درجة البكالوريوس فيها بالعام 2015, ومن ثم الماجستير بالعام الأكاديمي 2018، وحالياً أعمل في جامعة البعث (سورية) كمحاضر ، 

أعتقد أن العلم يجب أن يكون متاحاً للجميع ومجاني، وأحد أبرز مجالات اهتمامي هي طرق التعليم الحديثة التفاعلية

مجالات دراستي متنوعة , واركز عموماً على العلوم الهندسية , إلى جانب بعض الهوايات
